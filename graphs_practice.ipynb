```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# sample data
x_values = [1,2,3,4]
y_values = [10,20,30,45]
```

```python
# linegraph
sns.lineplot(x = x_values,y= y_values)
plt.xlabel('huuya')
plt.ylabel('kamlesh')
plt.title('Line plot using seaborn')
plt.grid(True)
plt.show()
```
